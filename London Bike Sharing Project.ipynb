```python
# importing pandas library
import pandas as pd
```

```python
# reading csv
bikes = pd.read_csv('london_merged.csv')
bikes.head()
```

```python
# exploring dataset
bikes.info()
```

```python
bikes.shape
```

```python
# counting unique values for weather code column
bikes.weather_code.value_counts()
```

```python
# counting unique values for season column
bikes.season.value_counts()
```

```python
# renaming column names
new_cols_dict ={
    'timestamp':'time',
    'cnt':'count', 
    't1':'temp_real_C',
    't2':'temp_feels_like_C',
    'hum':'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather'
}
bikes.rename(new_cols_dict, axis=1, inplace=True)
```

```python
bikes.head()
```

```python
# turning humidity values to percentages
bikes.humidity_percent = bikes.humidity_percent / 100
```

```python
bikes.head()
```

```python
# season dictionary mapping season column values to written season values
season_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter'
}

# weather dictionary mapping weather column values to written weather values
weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}

# converting season column datatype to string
bikes.season = bikes.season.astype('str')
# mapping season values
bikes.season = bikes.season.map(season_dict)

# converting season column datatype to string
bikes.weather = bikes.weather.astype('str')
# mapping weather values
bikes.weather = bikes.weather.map(weather_dict)

bikes.head()
```

```python
# exploring is_holiday & is_weekend values
```

```python
bikes['is_holiday'].value_counts()
```

```python
bikes['is_weekend'].value_counts()
```

```python
# converting is_holiday & is_weekend values from 1/0 to Yes/No
```

```python
# map for columns values
bool_dict = {'0.0':'No', '1.0':'Yes'}

# converting columns datatypes to str
bikes.is_holiday = bikes.is_holiday.astype('str')
bikes.is_weekend = bikes.is_weekend.astype('str')

# mapping columns values to Yes/No
bikes.is_holiday = bikes.is_holiday.map(bool_dict)
bikes.is_weekend = bikes.is_weekend.map(bool_dict)

bikes.head()
```

```python
# dropping duplicates if any
bikes.drop_duplicates(inplace=True)
```

```python
# saving final dataframe to excel sheet for visualization via Tableau
bikes.to_excel('london_bikes_final.xlsx', sheet_name='Data')
```
